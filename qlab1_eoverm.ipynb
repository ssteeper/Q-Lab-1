```python
import scipy.optimize as opt
```

```python
#all in mm
yoffset = array([-.5, -1, -1.5, -1.5, -2]);
xdata = array([20, 40, 60, 80, 100]);
ey= array([.5, .5, .5, .5, .5]);

y1u = [1, 3, 7, 13, 20];
y1d = [-2, -5, -10, -17, -26];

y15u = [1, 3, 7, 13, 20];
y15d = [-2, -5, -10, -16, -25];

y2u = [1, 3, 7, 12, 20];
y2d = [-2, -5, -10, -16, -24];

y25u = [1, 3, 7, 13, 20];
y25d = [-2, -5, -10, -16, -24];

y3u = [1, 3, 7, 13, 20];
y3d = [-2, -5, -10, -16, -24];

a = [y1u, y1d, y15u, y15d, y2u, y2d, y25u, y25d, y3u, y3d]
b=[];c=[]
#subtracting yoffset from each set
for i in a:
    b.append((array(i)-yoffset))
#averaging upwards and downwards data
ys = array([b[0]+abs(b[1]), b[2]+abs(b[3]), b[4]+abs(b[5]), b[6]+abs(b[7]), b[8]+abs(b[9])]) / 2.0

volts = 1000
for i in ys:
    print str(volts) + 'V:  ' + str(i)
    volts += 500
```

```python
def circle(x, r, x0, y0):
    return -sqrt(r**2-(x-x0)**2)+y0
```

```python
guess = array([250.0, -3, 250.0])
x=arange(0,120,.1)

params=[]
volts = 1000
for i in ys:
    ydata = i
    fit = opt.curve_fit(circle, xdata, ydata, guess, ey)
    R=fit[0][0];X0=fit[0][1];Y0=fit[0][2]
    dR = sqrt(fit[1][0][0])   #curve_fit outputs a 3x3 covariance matrix, dr is the sqrt of the 1,1 spot
    params.append([[R,X0,Y0], dR])
    
    fig = figure(figsize=(16, 10), dpi=100)
    errorbar(xdata,ydata,ey, c='black',fmt='o', markersize=5)
    plot(x, circle(x,R,X0,Y0), c='red')
    show()
```

```python
volts = 1000
for i in params:
    print str(volts) + 'V: [r,x0,y0] = ' + str(i[0]) + ', dr = ' + str(i[1])
    volts += 500
```
